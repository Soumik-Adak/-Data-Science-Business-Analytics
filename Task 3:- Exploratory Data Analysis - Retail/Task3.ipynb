```python
df1 = df.drop("Postal Code", axis=1)
df1 = df.drop("Country", axis=1)
```

# Visualize the relation between 'Sales' and 'Profit'

```python
plt.scatter(x = df1.Sales, y = df.Profit, )
plt.title("Sales vs Profit", fontweight = "bold", fontsize = 20)
plt.xlabel("Sales", fontsize = 15)
plt.ylabel("Profit", fontsize = 15)
plt.show()
```


# Lets Visualize the relationship between 'Discount' and 'Profit'

```python
plt.scatter(x = df1.Discount, y = df.Profit, )
plt.title("Discount vs Profit", fontweight = 'bold', fontsize = 20)
plt.xlabel("Discount", fontsize = 15)
plt.ylabel("Profit", fontsize = 15)
plt.show()
```

# Lets check trends of Sales for "Segment"

```python
plt.figure(figsize=(10,6))
sns.countplot(x = 'Segment', data = df1)
plt.show()
```

```python
df1['Segment'].value_counts().plot.pie(figsize = (6,6), autopct = '%.2f')
plt.title("Segment wise sales pie chart", fontweight = 'bold', fontsize = 15)
```

# Lets check trends of sales for "Region"

```python
plt.figure(figsize = (15 , 7))
sns.countplot(x = 'Region', data = df1)
plt.show()
```

```python
df1['Region'].value_counts().plot.pie(figsize = (6,6), autopct = '%.2f')
plt.title("Region wise sales pie chart", fontweight = 'bold', fontsize = 15)
```

# Check trends of sales for "Category"

```python
plt.figure(figsize = (15, 7))
sns.countplot(x = "Category", data = df1)
plt.show()
```

```python
df1['Category'].value_counts().plot.pie(figsize = (6,6), autopct = '%.2f')
plt.title("Category wise sales pie chart", fontweight = 'bold', fontsize = 15)
```

```python
plt.figure(figsize = (15,8))
sns.countplot(x = "Sub-Category", data = df1)
plt.xticks(rotation = (90))
plt.show()
```

```python
plt.figure(figsize = (15,10))
sns.countplot(x = "State", data = df1)
plt.xticks(rotation = (90))
plt.show()
```
